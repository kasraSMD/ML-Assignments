```python
from sklearn.metrics import f1_score, recall_score, precision_score, accuracy_score, confusion_matrix
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
  
# # metadata 
# print(heart_disease.metadata) 
  
# # variable information 
# print(heart_disease.variables) 
```

```python
X = X[['trestbps', 'chol', 'fbs', 'thalach', 'exang', 'oldpeak' ,'ca']]
```

```python
x_train, x_test, y_train, y_test = train_test_split( X, y, 
                                                    test_size=0.2, random_state=4)
```

```python
classifier_model = XGBClassifier()
classifier_model.fit(x_train, y_train)
y_pred = classifier_model.predict(x_test)
```

```python
print("accuracy score for x test prediction: ", accuracy_score(y_test, y_pred))
```

```python
results = []

learning_rates = [0.1, 0.01]
max_depths = [3, 6]
n_estimators = [100, 200]

for learning_rate in learning_rates:
    for max_depth in max_depths:
        for n_estimator in n_estimators:
            model = XGBClassifier(learning_rate=learning_rate, max_depth=max_depth, 
                                  n_estimators=n_estimator)
            model.fit(x_train, y_train)

            test_pred = model.predict(x_test)

            test_accuracy = accuracy_score(y_test, test_pred)

            results.append({
                'learning_rate': learning_rate,
                'max_depth': max_depth,
                'n_estimators': n_estimator,
                'accuracy': test_accuracy
            })
```

```python
for result in results:
    print("Learning Rate:", result['learning_rate'])
    print("Max Depth:", result['max_depth'])
    print("n_estimators:", result['n_estimators'])
    print("Accuracy:", result['accuracy'])
    print()
```
