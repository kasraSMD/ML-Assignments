```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dot, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import recall_score
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
# Load the dataset
!wget http://files.grouplens.org/datasets/movielens/ml-100k.zip
!unzip ml-100k.zip
ratings = pd.read_csv('ml-100k/u.data', sep='\t', header=None, names=['user_id', 'item_id', 'rating', 'timestamp'])
ratings.drop('timestamp', axis=1, inplace=True)
ratings.head()
```

```python
# Encode user and item ids
le = LabelEncoder()
ratings['user_id'] = le.fit_transform(ratings['user_id'])
ratings['item_id'] = le.fit_transform(ratings['item_id'])

# Prepare training and test data
train_data, test_data = train_test_split(ratings, test_size=0.2)
train_data.head()
```

```python
# Encode user and item ids
le = LabelEncoder()
ratings['user_id'] = le.fit_transform(ratings['user_id'])
ratings['item_id'] = le.fit_transform(ratings['item_id'])

# Prepare training and test data
train_data, test_data = train_test_split(ratings, test_size=0.2)
```

```python
# Number of users, movies, min and max ratings
num_users = len(ratings['user_id'].unique())
num_movies = len(ratings['item_id'].unique())
min_rating = ratings['rating'].min()
max_rating = ratings['rating'].max()

# Print dataset statistics
print("Number of users: {}, Number of Movies: {}, Min rating: {}, Max rating: {}".format(num_users, num_movies, min_rating, max_rating))
```

```python
# Define the model with dynamic Embedding Layers
embedding_size = 64
user_id_input = Input(shape=[1], name='user')
item_id_input = Input(shape=[1], name='item')

user_embedding_layer = Embedding(output_dim=embedding_size, input_dim=num_users, input_length=1, name='user_embedding')
item_embedding_layer = Embedding(output_dim=embedding_size, input_dim=num_movies, input_length=1, name='item_embedding')

user_embedding = user_embedding_layer(user_id_input)
item_embedding = item_embedding_layer(item_id_input)

user_vecs = Flatten()(user_embedding)
item_vecs = Flatten()(item_embedding)

y = Dot(1, normalize=False)([user_vecs, item_vecs])

model = Model(inputs=[user_id_input, item_id_input], outputs=y)
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
# Train the model
model.fit([train_data['user_id'], train_data['item_id']], train_data['rating'], batch_size=64, epochs=5, validation_split=0.1)
```

```python
# Plot the training history
history = model.history.history
plt.plot(history['loss'], label='Training Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
# Calculate Recall@10 on test data
def calculate_recall(model, test_data, k=10):
    user_ids = test_data['user_id'].unique()
    recalls = []
    for user_id in user_ids:
        user_data = test_data[test_data['user_id'] == user_id]
        user_movies = user_data['item_id'].values
        user_ratings = user_data['rating'].values
        predictions = model.predict([tf.constant([user_id] * num_movies), tf.constant(range(num_movies))])
        predicted_ratings = predictions.flatten()
        top_k_indices = predicted_ratings.argsort()[-k:][::-1]
        top_k_movies = [movie_id for movie_id in top_k_indices if movie_id not in user_movies][:k]
        true_positives = sum([1 for movie_id in top_k_movies if movie_id in user_movies])
        recall = true_positives / len(user_movies)
        recalls.append(recall)
    mean_recall = sum(recalls) / len(recalls)
    return mean_recall

recall_at_10 = calculate_recall(model, test_data)
print(f'Recall@10 on test data: {recall_at_10}')
```

```python
def recommend_movies(model, user_id, k=10):
    movie_ids = list(range(num_movies))
    user_ids = [user_id] * num_movies

    predictions = model.predict([tf.constant(user_ids), tf.constant(movie_ids)])
    predicted_ratings = predictions.flatten()

    top_k_indices = predicted_ratings.argsort()[-k:][::-1]

    recommended_movies = [movie_id for movie_id in top_k_indices]
    return recommended_movies
```

```python
# Recommend movies for a specific user
user_id_to_recommend = 0  # Change to the desired user id
recommended_movies = recommend_movies(model, user_id_to_recommend)
print(f'Recommended movies for user {user_id_to_recommend}: {recommended_movies}')
```
